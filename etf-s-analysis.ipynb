In this notebook we will analayze the performance of four big ETFs (Exchange Traded Fund), through the Covid-19 period. Each of the funds represent a differenf stocks sector:

QQQ - is a fund that follow the Nasdaq 100 index which includes big tech stocks.
VOO - is a fund that follow the S&P 500 index which includes the bigest 500 stocks traded in the US.
XLE - is a fund that follow the energy sector index.
JETS - is a fund that follow the aviation sector index.

As we shall see, there is a huge difference in the performance of those ETFs through the Covid-19 period.

```python
v = yf.HistoricalPrices('voo', '2019-10-21', '2020-10-21')
df_voo = v.to_dfs()['Historical Prices']
df_voo['Name'] = 'VOO'

df_voo.head()
```

```python
x = yf.HistoricalPrices('xle', '2019-10-21', '2020-10-21')
df_xle = x.to_dfs()['Historical Prices']
df_xle['Name'] = 'XLE'

df_xle.head()
```

```python
j = yf.HistoricalPrices('jets', '2019-10-21', '2020-10-21')
df_jets = j.to_dfs()['Historical Prices']
df_jets['Name'] = 'JETS'

df_jets.head()
```

# 4. Daily return for each of the ETFs

Here we will analayze the daily changes for each ETF through the last year

# 5. Create correlation heatmap of the 4 ETFs

```python
close_prices = pd.DataFrame([df_qqq['Adj Close'], df_voo['Adj Close'], df_xle['Adj Close'], df_jets['Adj Close']])

close_prices = close_prices.transpose()

close_prices.columns = ['QQQ','VOO','XLE','JETS']

fig, ax = plt.subplots(figsize=(5,5))

corr = close_prices.corr()

sns.heatmap(corr, annot=True, cmap='YlGnBu', ax = ax)
```

There is a strong correlation between QQQ and VOO. A possible explanation is that there is some stocks that participate in both funds (actually all of the tech giant such as Apple, Amazone, Facebook etc, are in both)

However, the correlation of XLE and JETS is even stronger. A possible explanation is that with the freezing of flights following the Covid-19 pandemic the two sectors were equally affected.

Now lets dive deeper and draw the correlation lines for each ETFs pair:

```python
g = sns.PairGrid(close_prices, diag_sharey=False, corner=True, height=2.5)
g.map_lower(sns.regplot, color="#930096")
g.map_diag(sns.kdeplot)
```

# 6. Calculating Bollinger Bands for each of the ETFs

Bollinger Bands is a bands that stand in a distance of 2 STD from the moving average (usualy MA of the last 20 days). This is a powerful tool in technical analysis.

```python
for etf in etfs:
    
    etf['MA20'] = etf['Adj Close'].rolling(window=20).mean()
    etf['20dSTD'] = etf['Adj Close'].rolling(window=20).std() 

    etf['Upper'] = etf['MA20'] + (etf['20dSTD'] * 2)
    etf['Lower'] = etf['MA20'] - (etf['20dSTD'] * 2)
```

```python
plt.figure(figsize=(12, 10))

for i, etf in enumerate(etfs, 1):
    plt.subplot(2, 2, i)
    etf['Adj Close'].plot(linewidth=2)
    etf['Lower'].plot(linewidth=2)
    etf['MA20'].plot(linewidth=2)
    etf['Upper'].plot(linewidth=2)
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"{etf['Name'][i - 1]}")
    plt.legend()
```

# 7. Predict future prices for VOO ETF

Lets try now to build a model that predict future prices for the VOO ETF. We will use the Random Forest Regressor for that challenge.


```python
df_voo_close = df_voo[['Adj Close']]
df_voo_close.head(4)
```

```python
# Create a variable to predict 'x' days out into the future

future_days = 35

# Create a new column (the target or dependent variable) shifted 'x' units/days up

df_voo_close['Prediction'] = df_voo_close[['Adj Close']].shift(-future_days)
df_voo_close.tail()
```

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor

# train the model
voo_train_set = df_voo_close[0:-future_days]
x_train = np.array(voo_train_set.drop(['Prediction'],1))
y_train = np.array(voo_train_set.drop(['Adj Close'],1))
tree = DecisionTreeRegressor().fit(x_train, y_train)
```

```python
#predict future values

x_future = df_voo_close.drop(['Prediction'],1)
x_future = x_future.tail(future_days)
x_future = np.array(x_future)

tree_prediction = tree.predict(x_future)
print(tree_prediction)
```

```python
#add predictions to df

df_voo_close['Tree Prediction'] = np.nan
df_voo_close['Tree Prediction'][-future_days:] = tree_prediction

df_voo_close.tail()
```

```python
#visualize the results

plt.figure(figsize=(10,6))

df_voo_close['Adj Close'].plot(linewidth=2)
df_voo_close['Tree Prediction'].plot(linewidth=2)
plt.ylabel('Adj Close USD ($)')
plt.title('Decision Tree Regressor Model')

plt.legend()
plt.show()
```
